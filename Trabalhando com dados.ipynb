```python
#Qual Ã© a coluna de salario base?
#R:BasePay
bdgeral.BasePay.tolist()
```

```python
#Qual Ã© o salÃ¡rio mÃ©dio?

SomaTotal = int(bdgeral.BasePay.sum())
Divisao = 5 

print('R$: ',SomaTotal/Divisao)
```

```python
#Qual Ã© o salÃ¡rio mÃ¡ximo?

MaiorSalario = bdgeral['BasePay'].max()
print("R$: ", MaiorSalario)
```

```python
#Qual Ã© o emprego do JOSEPH DRISCOLL? EmployeeName e JobTitle

bdgeral.loc[bdgeral['EmployeeName']=='JOSEPH DRISCOLL']['JobTitle']
```

```python
# Qual Ã© o salÃ¡rio mÃ©dio por ano?

Ano2011 = bdgeral.loc[bdgeral['Year']==2011]['BasePay'].sum()/12
Ano2012 = bdgeral.loc[bdgeral['Year']==2012]['BasePay'].sum()/12
Ano2013 = bdgeral.loc[bdgeral['Year']==2013]['BasePay'].sum()/12
Ano2014 = bdgeral.loc[bdgeral['Year']==2014]['BasePay'].sum()/12



print('A mÃ©dia do salario de 2011 Ã©: R$: ',Ano2011)
print('A mÃ©dia do salario de 2011 Ã©: R$: ',Ano2012)
print('A mÃ©dia do salario de 2011 Ã©: R$: ',Ano2013)
print('A mÃ©dia do salario de 2011 Ã©: R$: ',Ano2014)
```
